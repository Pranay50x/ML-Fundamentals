<b>Scatter plot</b><br>
<p>
The scatter plot is a graph in which the values of two variables are plotted along two axes, the pattern of the resulting points revealing any correlation present. The scatter plot is a useful tool for visualizing the relationship between two variables.
</p>
<p>Here we take random values from the numpy library and plot them using the scatter plot.</p>

<b>Syntax: `plt.scatter(x, y)`</b>
where x and y are the values of the two variables to be plotted.<br>

<b>Attributes with the Scatter Plot</b>
1. `c`: The color of the dots in the scatter plot.<br>
2. `marker`: The marker can be used instead of the default dots<br>
3. `s`: The size of the dots in the scatter plot.<br>
4. `alpha`: The transparency of the dots in the scatter plot.(when points are coinciding we can use this to distinguish b/w them)<br>

```python
x_data = np.random.rand(100)*500 
y_data = np.random.rand(100)*500

plt.scatter(x_data, y_data,c='red',marker='*',s=100, alpha=0.5)
plt.show()
```

<b>Bar Plot</b>
<p>
A bar plot is a plot that presents categorical data with rectangular bars with lengths proportional to the values that they represent. A bar plot shows comparisons among discrete categories. One axis of the plot shows the specific categories being compared, and the other axis represents a measured value.</p>
Here we try to show you by using lists in python.<br>

<b>Syntax: `plt.bar(x, y)`</b><br>
where x and y are the values of the two variables to be plotted.<br>

```python
import matplotlib.ticker as ticker #type: ignore

singers = ['The Weekend', 'Taylor Swift', 'Billie Eilish', 'Post Malone', 'Sabrina Carpenter']
listeners = [106012626, 98370718, 98245284, 87826851, 85223493] #number of monthly listeners on Spotify

plt.bar(singers, listeners, color='lime', width=0.6,edgecolor='black',linewidth=2)

# Format y-axis as 'Millions' and add thousands separator
formatter = ticker.FuncFormatter(lambda x, pos: '{:,.0f}M'.format(x/1e6))   
plt.gca().yaxis.set_major_formatter(formatter)

plt.xticks(rotation=45) # Optional: Rotate x-axis labels for better readability
plt.title('Monthly Listeners of Top 5 Singers on Spotify')
plt.xlabel('Singers')
plt.ylabel('Number of Monthly Listeners')
plt.show()
```

<b>Box Plot</b>
<p>
A box plot is a method for graphically depicting groups of numerical data through their quartiles. The box extends from the Q1 to Q3 quartile values of the data, with a line at the median (Q2). The whiskers extend from the edges of the box to show the range of the data. The position of the whiskers is set by default to 1.5 * IQR (IQR = Q3 - Q1) from the edges of the box, but can be set to any other value.</p>
Here we try to show you by using random values from the numpy library.<br>

<b>Syntax: `plt.boxplot(x)`</b><br>
where x is the values of the variable to be plotted.<br>


```python
ar1 = np.random.normal(170,10,250)

plt.boxplot(ar1)
```

This code snippet generates and visualizes a dataset using NumPy and Matplotlib:

1. **Data Generation with NumPy**:
   - `np.random.normal(loc, scale, size)`: Draws random samples from a normal (Gaussian) distribution.
     - `a1`: 200 samples from a normal distribution with a mean (`loc`) of 100 and a standard deviation (`scale`) of 10.
     - `a2`: 200 samples from a normal distribution with a mean of 90 and a standard deviation of 20.
     - `a3`: 200 samples from a normal distribution with a mean of 80 and a standard deviation of 30.
     - `a4`: 200 samples from a normal distribution with a mean of 70 and a standard deviation of 40.

2. **Data Concatenation**:
   - `np.concatenate()`: Merges the four arrays (`a1`, `a2`, `a3`, `a4`) into a single array named `data`. This results in a combined dataset of 800 samples, showcasing a mix of distributions with varying means and standard deviations.

3. **Visualization with Matplotlib**:
   - `plt.boxplot(data)`: Creates a box plot for the concatenated data. A box plot visually shows the distribution of the numerical data and its skewness through displaying the data quartiles (or percentiles) and averages. This plot will help in understanding the combined distribution of the four normally distributed datasets, including their median, range, and any potential outliers.

This approach is useful for analyzing the aggregate behavior of multiple datasets with different normal distributions in a single visualization.

```python
a1 = np.random.normal(100, 10, 200)
a2 = np.random.normal(90, 20, 200)
a3 = np.random.normal(80, 30, 200)
a4 = np.random.normal(70, 40, 200)

data = np.concatenate((a1,a2,a3,a4))
plt.boxplot(data)
```

In this code snippet, we're generating and visualizing data using NumPy and Matplotlib:

1. `np.linspace(start, stop, num)`: Generates `num` evenly spaced samples, calculated over the interval `[start, stop]`. We create four such linearly spaced arrays:
   - `l1`: From 0 to 10, with 100 samples.
   - `l2`: From 10 to 20, with 100 samples.
   - `l3`: From 20 to 30, with 100 samples.
   - `l4`: From 30 to 40, with 100 samples.

2. `np.concatenate()`: Combines the four arrays (`l1`, `l2`, `l3`, `l4`) into a single array named `data`.

3. `plt.boxplot(data)`: Plots a box plot using Matplotlib for the combined data. A box plot is a standardized way of displaying the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It can tell you about your outliers and what their values are. It can also tell you if your data is symmetrical, how tightly your data is grouped, and if and how your data is skewed.

This process results in a single box plot that visualizes the distribution of the concatenated data from `l1`, `l2`, `l3`, and `l4`.

```python
l1 =  np.linspace(0 , 10, 100)
l2 = np.linspace(10, 20, 100)
l3 = np.linspace(20, 30, 100)
l4 = np.linspace(30, 40, 100)

data = np.concatenate((l1,l2,l3,l4))

plt.boxplot(data)

```

<b>3D Line Plot</b>

```python
x= np.arange(-5, 5, 0.1)
y= np.arange(-5, 5, 0.1)
z = np.cos(np.sqrt(x**2 + y**2))

ax = plt.axes(projection='3d')
ax.scatter(x, y, z, c=z, cmap='magma', linewidth=0.2)
plt.title('3D Scatter Plot')
plt.show()
```

<b>3D Surface Plot</b>

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='inferno')
ax.view_init(azim=120, elev=30)
plt.title('3D Surface Plot')
plt.show()
```

<b>Animated Plot</b><br>
Animated plots are used to visualize data that changes over time. They are useful for showing trends, patterns, and relationships in time-series data. In this code snippet, we're creating an animated plot using Matplotlib and NumPy to visualize the movement of a point over time.


```python
from matplotlib.animation import FuncAnimation #type: ignore

fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'ro', animated=True)

def init():
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1, 1)
    return ln,

def update(frame):
    xdata.append(frame)
    ydata.append(np.sin(frame))
    ln.set_data(xdata, ydata)
    return ln,

ani = FuncAnimation(fig, update, frames=np.linspace(0, 2*np.pi, 128), init_func=init, blit=True)


ani.save('sine_wave_animation.gif', writer='imagemagick')

plt.show()
```
